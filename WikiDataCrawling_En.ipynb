# Crawling data from wikipedia in English

```python
# english wiki page crawling

def wikiDataCrawling_En(url, termname):
    #English 

    url = url
    soup = BeautifulSoup(urllib.request.urlopen(url), 'html.parser')
    data = list()
    
    #define english noise headline/paragraph(with no/little context) list, this list needs to be updated as more pages are crawled
    noise_english = [ "Historiography","Audiobooks","Lists","Other","Personal tools","Namespaces",
 "\nVariants\n","Views", "More", "\nSearch\n", "Navigation", "Interaction", "Tools", "Print/export", "In other projects", "Languages","Official websites","Academic",
 "Reflections on the Bush presidency","See also[edit]", "Filmography","Sources","Bibliography[edit]", "Informational notes[edit]","References in popular culture[edit]",
 "Works[edit]","Honors[edit]", "Filmography[edit]","Missed roles[edit]","Awards and nominations[edit]","See also[edit]","Performances and awards[edit]"
  "Primary sources","","Official","Speeches and statements","Other","Footnotes[edit]","References[edit]","Further reading[edit]","External links[edit]",'Media','Honorary degrees', 'Footnotes','Notes', 'References', 'Bibliography','Books written','Contents', 'See also', 'Notes and references', 'Further reading', 'External links', 'Navigation menu','Electoral history','Media coverage']
    
    #a = -1 #count list length
    
    #remove noise tables
    for s in soup.body.findAll('table'):
        s.decompose()
    
    #get the wikipedia page body text
    for row in soup.body.findAll(['h1','h2','p']):
        if row is not '':
        #if len(row.text.split())>1 :
            data.append(row.text)
            #a += 1

    #remove noise headlines
    for n in noise_english:
        if n in data:
            data.remove(n)

    #get the headlines         
    headline = list()
    for row in soup.body.findAll(['h1','h2']):
        if row is not None:
            
        #if len(row.text.split())>1 :
            headline.append(row.text)

    #remove noise headlines
    for n in noise_english:
        if n in headline:
            headline.remove(n)

    print (json.dumps(headline,indent=1))

    #get the headline index        
    headlineIndex = list()        
    for h in headline:
        headlineIndex.append(data.index(h))
    #print(headlineIndex)

    ##get the context for each headline
    p = list()
    i = 0
    for index in range(len(headlineIndex)-1): #this can only reach to the penultimate headline
        p.append(data[headlineIndex[index]+1: headlineIndex[index+1]])

    #get the context of the last headline    
    p.append(data[headlineIndex[-1]+1: len(data)])


    #conbine the headline and context into a dictionary

    english_headline_content_dict = {}
    for i in range(len(headline)):
        english_headline_content_dict[headline[i]] = p[i]
    
    #print(dic)

    #json.dumps(data)
    #To check the result just print the above line or assign it variable and print 

    #print (json.dumps(english_headline_content_dict,indent=1))

    with open("source_en_"+ termname+".json", 'w') as f:
         json.dump(english_headline_content_dict, f)

    #print(a)
    #return english_headline_content_dict, headline
```
