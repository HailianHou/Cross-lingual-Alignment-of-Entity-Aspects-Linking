```python
def wikiDataCrawling_De(url, termname, noise):
    #German

    url = url
    #page = urllib.request.urlopen(url)
    soup = BeautifulSoup(urllib.request.urlopen(url), 'html.parser')

    data = list()
    #define germany noise headline (with no/little context) list
    noise_german = noise

    #a = -1 #count list length

    #remove noise tables
    for s in soup.body.findAll('table'):
        s.decompose()

    #get the wikipedia page body text
    for row in soup.body.findAll(['h1','h2','p']):
        data.append(row.text)
            #a += 1

    #remove noise headlines
    for n in noise_german:
        if n in data:
            data.remove(n)

    #get the headlines         
    headline = list()
    for row in soup.body.findAll(['h1','h2']):
        headline.append(row.text)

    #remove noise headlines
    for n in noise_german:
        if n in headline:
            headline.remove(n)

    print (json.dumps(headline,indent=1))

    #get the headline index in data list        
    headlineIndex = list()        
    for h in headline:
        headlineIndex.append(data.index(h))
    #print(headlineIndex)

    ##get the context for each headline
    p = list()
    i = 0
    for index in range(len(headlineIndex)-1): #this can only reach to the penultimate headline
        p.append(data[headlineIndex[index]+1: headlineIndex[index+1]])

    #get the context of the last headline     
    p.append(data[headlineIndex[-1]+1: len(data)])


    #conbine the headline and context into a dictionary

    german_headline_content_dict = {}
    for i in range(len(headline)):
        german_headline_content_dict[headline[i]] = p[i]

    #print(dic)

    #json.dumps(data)
    #To check the result just print the above line or assign it variable and print 

    #print (json.dumps(german_headline_content_dict,indent=1))
    #return headline, german_headline_content_dict

    with open("source_de_"+ termname+".json", 'w') as f:
         json.dump(german_headline_content_dict, f)

    
    #return german_headline_content_dict, headline
    #print(a)
```

```python
# get the link of wikipages in german and chinese

def getLanguageLink(url):    
    
    url = url 

    soup = BeautifulSoup(urllib.request.urlopen(url), 'html.parser')

    #get other language links of Barack_Obama
    languageList = list()
    for link in soup.findAll('a', lang=['de','zh']):
        languageList.append(link.get('href'))
        #print(link.get('href'))

    return languageList
```

# Translation from deutsch to english

```python
from googletrans import Translator
from collections import defaultdict
```

```python
def translate(text, headline, src, dest, termname):
    text = text
    dest = dest
    translator = Translator()
    dict_trans = defaultdict(list)
    for el in headline:
        h = translator.translate(text = el, dest = dest).text
        t = translator.translate(text = text[el], dest = dest)
        for m in t:
            dict_trans[h].append(m.text)
            #print(m.text)

    print(dict_trans)
    with open('translation_'+ src +"_to_"+ dest+'_'+termname+".json", 'w') as f:
        json.dump(dict_trans, f)
    
    return dict_trans
```
