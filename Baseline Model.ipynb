```python
import pandas as pd
import json
```

```python
####get the baseline run file for each en-de wikipage pair

def get_baseline_run(l, m, name, trs, scr):
    
    l_len = len(l)
    m_len = len(m)

    #en_to_de_run_obama = list()
    r = []

    if l_len <= m_len:
        print("length l is smaller than length m")
        for i in range(l_len):
            j= i
            q = m_len
            if j == 0:
                for x in range(m_len):
                    r.append([trs+'_'+name + "_h"+ str(i), "0", scr+ '_'+name + "_h"+str(x), m_len+1-q, q/m_len, "baseline"])
                    q -= 1
            else:
                q = m_len
                while j < i+m_len:
                    while j < m_len:
                        r.append([trs+'_' +name + "_h"+str(i), "0", scr+ '_'+name + "_h"+str(j), m_len+1-q, q/m_len, "baseline"])
                        #list.append(l[i], " 0 ", m[j], q, q/m_len, "method")
                        j += 1
                        q -= 1
                    else:
                        p = j-m_len
                        r.append([trs+'_'+name + "_h"+str(i), '0', scr+ '_'+name + "_h"+str(p), m_len+1-q, q/m_len, "baseline"])
                        #list.append(l[i], ' 0 ', m[j-m_len], q, q/m_len, "method")
                        j += 1
                        q -= 1
    else:
        print("length l is bigger than length m")
        for i in range(l_len):
            j = i
            q = m_len
            if i < m_len:
                if j == 0:
                    for x in range(m_len):
                        r.append([trs+ '_'+name + "_h"+ str(i), "0", scr+ '_'+name + "_h"+str(x), m_len+1-q, q/m_len, "baseline"])
                        q -= 1
                else:
                    q = m_len
                    while j < i+m_len:
                        while j < m_len:
                            r.append([trs+ '_'+ name + "_h"+ str(i), "0", scr+ '_'+name + "_h"+str(j), m_len+1-q, q/m_len, "baseline"])
                            j += 1
                            q -= 1
                        else:
                            p = j-m_len
                            r.append([trs+ '_'+name + "_h"+ str(i), '0', scr+ '_'+name + "_h"+str(p), m_len+1-q, q/m_len, "baseline"])
                            j += 1
                            q -= 1
            else:
                q = m_len
                e = m_len - 1
                while e == m_len-1:
                    r.append([trs+ '_'+name + "_h"+ str(i), '0', scr+ '_'+name + "_h"+str(e), m_len+1-q, q/m_len, "baseline"])
                    e -=1
                    q -=1
                else:                    
                    while e>0:
                        s = m_len-(e+1)
                        r.append([trs + '_'+name + "_h"+ str(i), '0', scr+ '_'+name + "_h"+str(s), m_len+1-q, q/m_len, "baseline"])
                        e -=1
                        q -=1

    df = pd.DataFrame(r)
    df.to_csv('Baseline_'+trs+'_'+scr+'.txt', header=None, index=None, sep=' ', mode='a')
    
    print(df)



#, columns=[trs,"zero", scr, 'linking sequence', 'relevance', 'model']
```

```python
### run baseline model for all of the entities!! 
#Language pair: English-German

# define the entity list
entity_list = ['United_States','China','France', 'Germany', 'Japan', 'Asia', 'Europe', 'Italy', 
               'United_Kingdom','Canada','Russia','India', 'Israel','Brazil','Philippines', 
               'New_York_City','London','Singapore','Barack_Obama', 'Donald_Trump','Dubai','Los_Angeles','Paris',
               'Chicago','Washington,_D.C.','San_Francisco',
               'Mumbai','Rome','Toronto','Philadelphia','Monaco','Tokyo','Amsterdam','Boston',
               'Barcelona','Peking']

for el in entity_list:
    
    #load data
    with open('/home/hahou/WikiDataCrawling/German Corpus/source_de_'+el+'.json') as json_data:
        source_de = json.load(json_data)

    with open('/home/hahou/WikiDataCrawling/English Corpus/source_en_'+el+'.json') as json_data:
        source_en = json.load(json_data)
    
    #get section headlines
    en_headline = list(source_en.keys())
    de_headline = list(source_de.keys())
    
    #run baseline model!!
    get_baseline_run(en_headline, de_headline, el, 'en', 'de')
```
