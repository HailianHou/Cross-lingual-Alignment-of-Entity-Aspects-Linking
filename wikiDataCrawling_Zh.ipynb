```python
import urllib
from bs4 import BeautifulSoup
import urllib.parse
import urllib.request
import json
```

```python
noise_Chinese =['注释[编辑]', '參看[编辑]','參见[编辑]','参見[编辑]','參見[编辑]','来源[编辑]','注解[编辑]', '注释[编辑]', '參考文獻[编辑]', '外部链接[编辑]', '參見', '註釋', '参考文献', '外部連結',
    '目录', '相关逸事[编辑]',"重要演讲[编辑]","参见[编辑]","参考文献[编辑]",'外部連結[编辑]', '导航菜单',"著作[编辑]", '外部产生原因有以下三大因素：\n', '自身产生原因有以下三大因素：\n']
```

```python
def wikiDataCrawling_zh(url, termname, noise):
    #Chinese

    url = url
    #page = urllib.request.urlopen(url)
    soup = BeautifulSoup(urllib.request.urlopen(url), 'html.parser')

    data = list()
    #define Chinese noise headline (with no/little context) list
    noise_Chinese = noise

    #a = -1 #count list length

    #remove noise tables
    for s in soup.body.findAll('table'):
        s.decompose()

    #get the wikipedia page body text
    for row in soup.body.findAll(['h1','h2','p']):
        data.append(row.text)
            #a += 1

    #remove noise headlines
    for n in noise_Chinese:
        if n in data:
            data.remove(n)

    #get the headlines         
    headline = list()
    for row in soup.body.findAll(['h1','h2']):
        headline.append(row.text)

    #remove noise headlines
    for n in noise_Chinese:
        if n in headline:
            headline.remove(n)

    print (headline)

    #get the headline index in data list        
    headlineIndex = list()        
    for h in headline:
        headlineIndex.append(data.index(h))
    #print(headlineIndex)

    ##get the context for each headline
    p = list()
    i = 0
    for index in range(len(headlineIndex)-1): #this can only reach to the penultimate headline
        p.append(data[headlineIndex[index]+1: headlineIndex[index+1]])

    #get the context of the last headline     
    p.append(data[headlineIndex[-1]+1: len(data)])


    #conbine the headline and context into a dictionary

    Chinese_headline_content_dict = {}
    for i in range(len(headline)):
        Chinese_headline_content_dict[headline[i]] = p[i]

    #print(Chinese_headline_content_dict)

    #json.dumps(data)
    #To check the result just print the above line or assign it variable and print 

    #print (json.dumps(Chinese_headline_content_dict,indent=1))
    #return headline, Chinese_headline_content_dict

    with open("source_zh_"+ termname+".json", 'w', encoding = 'utf-8') as f:
         json.dump(Chinese_headline_content_dict, f, ensure_ascii=False)

    
    #return Chinese_headline_content_dict, headline
    #print(a)
```

```python
# get the link of wikipages in german and chinese

def getLanguageLink(url):    
    
    url = url 

    soup = BeautifulSoup(urllib.request.urlopen(url), 'html.parser')

    #get other language links of Barack_Obama
    languageList = list()
    for link in soup.findAll('a', lang=['de','zh']):
        languageList.append(link.get('href'))
        print(link.get('href'))

    return languageList
```

```python
entry_list = ['Donald_Trump','Barack_Obama']
for entry in entry_list:
    language = getLanguageLink("https://en.wikipedia.org/wiki/"+entry)
    wikiDataCrawling_zh(language[1], entry, noise_Chinese)
```
